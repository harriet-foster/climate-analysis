# House price prediction

In this exercise, we are going to apply a regression modelling technique to house price prediction using a subset of the [California house price dataset](https://www.kaggle.com/camnugent/california-housing-prices). Our dataset contains 200 observations for housing blocks in California obtained from the 1990 census. The dataset contains columns:

1. `longitude`: A measure of how far west a house is; a higher value is farther west

2. `latitude`: A measure of how far north a house is; a higher value is farther north

3. `housing_median_age`: Median age of a house within a block; a lower number is a newer building

4. `total_rooms`: Total number of rooms within a block

5. `total_bedrooms`: Total number of bedrooms within a block

6. `population`: Total number of people residing within a block

7. `households`: Total number of households, a group of people residing within a home unit, for a block

8. `median_income`: Median income for households within a block of houses (measured in tens of thousands of US Dollars)

9. `median_house_value`: Median house value for households within a block (measured in US Dollars)

10. `ocean_proximity`: Location of the house w.r.t ocean/sea

In this example, we are going to create a regression model to predict `median_house_value`.

## Linear regression
Split the dataset into a suitable training, validation and test set.

```python
training = df.iloc[0:120, :].copy()
validation = df.iloc[0:50, :].copy()
testing = df.iloc[0:30, :].copy()
training.shape, validation.shape, testing.shape
```

Next, we will build a linear regression model to predict median house value from median income using the training set.

### Building a linear regression model using scikit-learn
Look at `sklearn.linear_model.LinearRegression()` here: https://scikit-learn.org/1.5/modules/generated/sklearn.linear_model.LinearRegression.html.


Set X to your chosen input variable (I suggest `median_income`) and y to the output variable (`median_house_value`). You will need to reshape your input array to ensure they are of size `(N, 1)` for sklearn.

Fit the LinearRegression model and print the coefficients of the model.

```python
# Set up X and y
X = training["median_income"]
y = training["median_house_value"]

X = X.iloc[:, ].values.reshape(-1, 1)      # This reshapes the array so the inputs are the correct size (N, 1)
y = y.iloc[:, ].values.reshape(-1, 1)


# Create your linear regression model and print the coefficients and intercept.

from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X, y)
print(f"Model score: {model.score(X, y)}") # coefficient of determination of prediction, R^2
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

What does your model predict is the average increase in median house value associated with a \$10,000 increase in median income?

```python
# using the model coefficients to predict, as the coefficient for the median income shows
#the expected change in house value for a one-unit ($10,000) increase in median income:

# so here for every $10,000 increase in income, median house value increases by $41,919.72
```

Plot the linear regression line on top of the training data. What does this suggest about the appropriateness of the model?

<details><summary><b>Click for suggestion </b></summary>To plot the linear model prediction, you can create a vector that covers the `x`-axis and estimate `y` at these points.</details>


```python
y_predicted = model.predict(X)

plt.scatter(X, y, alpha = 0.5, label = "Training Data")
plt.plot(X, y_predicted, color = "red", label = "Linear Regression")
plt.xlabel("Median Income ($, tens of thousands)")
plt.ylabel("Median House Value, $")
plt.legend()
plt.show()

# so model is roughly appropriate, but missing nuance of outliers and vague curve of data - more datapoints for lower X vals.
```

## Check validation dataset
The next step of our supervised learning recipe is validation. Predict the `median_house_value` on the the validation dataset. How can you check the performance?

<details><summary><b>Click for suggestion </b></summary>To check performance, what error metrics could you look at? What plot would you use to compare the true dataset against the predictions? </details>


```python
y_validated_predicted = model.predict(validation["median_income"].values.reshape(-1, 1))
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(validation["median_house_value"].values.reshape(-1, 1), y_validated_predicted)
print(f"MSE: {mse}")
# this means optimising to where the predicted beta value = 0
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(validation["median_house_value"].values.reshape(-1, 1), y_validated_predicted)
print(f"MAE: {mae}")
# less sensitive to outliers, which is useful here i.e. by how any $ are we off in our prediction
# MAE better suggestion, still shows high 'error'
```

```python
# plotting predicted values versus actual - how well do predictions align with reality?

plt.scatter(validation["median_house_value"].values.reshape(-1, 1), y_validated_predicted, alpha=0.5)
plt.xlabel("Actual Median House Value")
plt.ylabel("Predicted Median House Value")
plt.title("Predicted vs Actual Values")

y_val = validation["median_house_value"].values.reshape(-1, 1)
min_val = min(y_val.min(), y_validated_predicted.min())
max_val = max(y_val.max(), y_validated_predicted.max())
plt.plot([min_val, max_val], [min_val, max_val], color="red")

plt.show()
# points on the red line are a perfect fit
```

```python
# plotting residuals i.e. error versus predictions
residuals = y_val - y_validated_predicted

plt.scatter(y_validated_predicted, residuals, alpha=0.5)
plt.axhline(0, color='red')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residual Plot")
plt.show()
```

```python
plt.hist(residuals, bins=40)
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.title("Residual Distribution")
plt.show()
```

## Improving the model

Now fit a linear regression model to the training set of the form:

\begin{equation}
P_i = a + b_1 I_i + b_2 I_i^2
\end{equation}

where $P_i$ is median house price in block $i$ and $I_i$ is median house income of the same block.

```python
# Create a new column in pandas for median income squared
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


# create features
X_train_2 = pd.DataFrame({
    "I": training["median_income"],
    "I2": training["median_income"] ** 2
})


model2 = LinearRegression().fit(X_train_2, y)

print("Intercept (a):", model2.intercept_)
print("Coefficients (b1, b2):", model2.coef_)
print(f"Intercept: {model2.intercept_}")
```

Again plot the model fit for this model versus data. How does the fit compare with the linear model?

```python
# Sort values for a smooth curve
I_sorted = np.linspace(training["median_income"].min(),
                       training["median_income"].max(), 300)

# Model predictions on the grid
y_predicted_2 = model2.predict(X_train_2)


# Plot real data
plt.scatter(X, y,
            alpha=0.4, label="Training data")

# Plot model curve
plt.plot(X_train_2, y_predicted_2,
         color="red", linewidth=2, label="Quadratic fit")

plt.xlabel("Median Income, tens of thousands, $")
plt.ylabel("Median House Value, $")
plt.title("Quadratic Model Fit vs Training Data")
plt.legend()
plt.show()
```
